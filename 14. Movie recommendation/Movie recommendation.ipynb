### Cosine Similarity

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies = pd.read_csv(r"D:\Python\ML projects\14. Movie recommendation\movies.csv")
```

```python
print(movies.shape)
movies.head()
```

```python
movies.info()
```

```python
# feature selection

selected_features = ['genres','keywords',str('vote_average'),'tagline','cast','director'] 
```

```python
# replacing the missing values with null string

for feature in selected_features:
    movies[feature] = movies[feature].fillna('')
```

```python
movies.isnull().sum()
```

```python
# combingin selected features

combined_features = movies['genres'] +' '+ movies['keywords'] +' '+ movies['tagline'] \
+" "+ movies['cast'] +' '+ movies['director']
```

```python
print\
(combined_features)
```

```python
#convert text into feature vectors using TfIdf

vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

```python
# finding similarity score using cosine similarity

similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
print('\n Shape is',similarity.shape)
```

```python
# getting movie name frmo user

user_movie = input('Enter your favorite movie name: ')
```

```python
# creating a list with all the movies given in the dataset

list_of_all_titles = movies['title'].tolist()
print(list_of_all_titles)
```

```python
# finding the close match for the movie name given by the user using difflib

find_close_match = difflib.get_close_matches(user_movie, list_of_all_titles)
print(find_close_match)
```

```python
# take the first value of find_close_match list

close_match = find_close_match[0]
print(close_match)
```

```python
# finding index of the movei with title

index_of_the_movie = movies[movies.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
# getting a list of similar movies based on index value

similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)

# The movies that would be similar to the given movie, will have higher similarity score
```

```python
len(similarity_score)
```

```python
# sorting movies based on similairty score

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse=True)
print(sorted_similar_movies)
```

```python
# take 5 movies from the above list and print it

print('Movies for you: \n')

i=1

for movie in sorted_similar_movies:
    index=movie[0]
    title_from_index = movies[movies.index == index]['title'].values[0]
    if (i<6):
        print(i,'.',title_from_index)
        i+=1
```

### Movie Recommendation System

```python
user_movie = input('Enter your favorite movie name: ')

list_of_all_titles = movies['title'].tolist()

find_close_match = difflib.get_close_matches(user_movie, list_of_all_titles)

close_match = find_close_match[0]

index_of_the_movie = movies[movies.title == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_the_movie]))

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse=True)

print('Movies for you: \n')

i=1

for movie in sorted_similar_movies:
    index=movie[0]
    title_from_index = movies[movies.index == index]['title'].values[0]
    if (i<6):
        print(i,'.',title_from_index)
        i+=1
```
